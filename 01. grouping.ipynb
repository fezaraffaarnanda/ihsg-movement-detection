```python
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/00. Drive PC/1.STIS/SEMNAS 2024/ihsg detik.xlsx")
df
```

```python
columns = ["Judul","Tanggal","Link","Teks Berita"]

df.columns=columns

df
```

```python
from datetime import datetime

# Fungsi untuk mengonversi string tanggal ke datetime
def convert_tanggal(tanggal_str):
    bulan_dict = {
        'Januari': 1, 'Februari': 2, 'Maret': 3, 'April': 4, 'Mei': 5, 'Juni': 6,
        'Juli': 7, 'Agustus': 8, 'September': 9, 'Oktober': 10, 'November': 11, 'Desember': 12,
        'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
        'Jul': 7, 'Aug': 8, 'Agu':8,'Sep': 9, 'Oct': 10, 'Okt':10,'Nov': 11, 'Dec': 12, 'Des':12
    }
    hari, bulan, tahun, waktu = tanggal_str.split()
    bulan_num = bulan_dict[bulan]
    jam, menit = waktu.split(':')
    return datetime(int(tahun), bulan_num, int(hari), int(jam), int(menit))

# Hapus spasi di awal
df['Tanggal'] = df['Tanggal'].str.strip()

# Konversi ke datetime dengan fungsi konversi
df['Tanggal'] = df['Tanggal'].apply(convert_tanggal)

# Group by date and aggregate the titles
grouped = df.groupby(df['Tanggal'].dt.date)['Teks Berita'].apply(list).reset_index()

# Buat DataFrame dengan struktur yang diinginkan
max_len = grouped['Teks Berita'].str.len().max()
new_data = {'Tanggal': grouped['Tanggal']}

for i in range(max_len):
    new_data[f'Teks Berita {i+1}'] = grouped['Teks Berita'].apply(lambda x: x[i] if i < len(x) else None)

result = pd.DataFrame(new_data)
```
